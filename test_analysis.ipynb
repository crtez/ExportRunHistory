```python
import pandas as pd
```

```python
# insert your json path between the parentheses like pd.read_json("C:\Users\admin\output\output.json")
df = pd.read_json()
```

```python
final_df = pd.DataFrame()
for i in range(len(df)):
    row = pd.DataFrame.from_dict(df.iloc[i][0], orient = "index").T #flattens each json object (each object is a run)
    final_df = final_df.append(row, ignore_index=True)
```

```python
def time_to_ms(time):
    min_ms = float(time[:2]) * 60000
    sec_ms = float(time[3:8]) * 1000
    return min_ms + sec_ms
```

```python
final_df['time_ms'] = final_df['time'].apply(lambda x: time_to_ms(x))
final_df
```

```python
# finds successful runs
cleared = final_df[final_df['result'] == 'RunHistoryScreen_Cleared']
display(cleared)
```

```python
mean_clear_time = cleared['time_ms'].mean() / 1000 / 60
print(mean_clear_time) # in minutes
```
